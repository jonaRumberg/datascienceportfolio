```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,accuracy_score, confusion_matrix, classification_report, accuracy_score, f1_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.svm import SVC, LinearSVC
from scipy.stats import chi2_contingency

df_train = pd.read_csv("data/smoker_train.csv")
df_train = df_train.drop_duplicates()
```

```python
df_with_duplicates = pd.read_csv("data/smoker_train.csv")

# Zeilenanzahl beider DataFrames
rows_train = len(df_train)
rows_with_duplicates = len(df_with_duplicates)

# Differenz berechnen
difference = rows_with_duplicates - rows_train

# Ausgabe
print(f"Anzahl Zeilen in df_train: {rows_train}")
print(f"Anzahl Zeilen in df_with_duplicates: {rows_with_duplicates}")
print(f"Differenz: {difference} Zeile(n)")
```

```python
# Übersicht über potenziell problematische Werte in df_train
# Du kannst alle Prüfungen beliebig erweitern oder anpassen

# 1. Altersprüfung (z. B. 20–90 Jahre sinnvoll)
print("Unplausible Werte – Alter:")
print(df_train[(df_train["age"] < 15) | (df_train["age"] > 90)])

# 2. Körpergröße (in cm)
print("Unplausible Werte – Körpergröße:")
print(df_train[(df_train["height(cm)"] < 130) | (df_train["height(cm)"] > 210)])

# 3. Körpergewicht (in kg)
print("Unplausible Werte – Körpergewicht:")
print(df_train[(df_train["weight(kg)"] < 35) | (df_train["weight(kg)"] > 200)])

# 4. Taillenumfang (cm)
print("Unplausible Werte – Taillenumfang:")
print(df_train[(df_train["waist(cm)"] < 40) | (df_train["waist(cm)"] > 160)])

# 5. Sehschärfe (typisch: 0.1 bis 2.0 – Werte über 2 sehr selten)
print("Unplausible Werte – Sehschärfe:")
print(df_train[(df_train["eyesight(left)"] < 0.1) | (df_train["eyesight(left)"] > 2.0) |
               (df_train["eyesight(right)"] < 0.1) | (df_train["eyesight(right)"] > 2.0)])

# 6. Blutdruck
print("Unplausible Werte – systolisch:")
print(df_train[(df_train["systolic"] < 70) | (df_train["systolic"] > 250)])
print("Unplausible Werte – diastolisch:")
print(df_train[(df_train["relaxation"] < 40) | (df_train["relaxation"] > 150)])

# 7. Blutwerte (hier beispielhafte Grenzen)
print("Unplausible Werte – Cholesterin:")
print(df_train[(df_train["Cholesterol"] < 50) | (df_train["Cholesterol"] > 400)])

print("Unplausible Werte – HDL:")
print(df_train[(df_train["HDL"] < 10) | (df_train["HDL"] > 150)])

print("Unplausible Werte – Triglyceride:")
print(df_train[(df_train["triglyceride"] < 10) | (df_train["triglyceride"] > 1000)])

print("Unplausible Werte – LDL:")
print(df_train[(df_train["LDL"] < 10) | (df_train["LDL"] > 300)])

print("Unplausible Werte – Fasting Blood Sugar:")
print(df_train[(df_train["fasting blood sugar"] < 50) | (df_train["fasting blood sugar"] > 300)])

# 8. Hämoglobin (z. B. 10–20 g/dl)
print("Unplausible Werte – Hämoglobin:")
print(df_train[(df_train["hemoglobin"] < 8) | (df_train["hemoglobin"] > 20)])

# 9. Serumkreatinin (z. B. 0.4–2.0 mg/dl)
print("Unplausible Werte – Serumkreatinin:")
print(df_train[(df_train["serum creatinine"] < 0.2) | (df_train["serum creatinine"] > 3.0)])

# 10. Leberwerte (AST, ALT, GTP – z. B. < 200 plausibel)
print("Unplausible Werte – Leberenzyme:")
print(df_train[(df_train["AST"] > 250) | (df_train["ALT"] > 250) | (df_train["Gtp"] > 350)])

# 11. Urinprotein (z. B. Werte 1–6 codiert)
print("Unplausible Werte – Urinprotein:")
print(df_train[~df_train["Urine protein"].isin([1, 2, 3, 4, 5, 6])])

# 12. Hearing, dental caries, smoking (erlaubte Werte: 0/1 oder 1/2)
print("Unplausible Werte – Nominalvariablen:")
for var in ["hearing(left)", "hearing(right)", "dental caries", "smoking"]:
    print(f"{var}: {df_train[var].unique()}")
```
